```python
import requests
import json
```

Configurar el endpoint de entrada de la API

```python
API_URL = 'http://163.117.129.205/api'
```

Obtener una serie de lemas. Se pueden utilizar los parámetros skip y limit para acceder a distintas palabras

```python
url = f'{API_URL}/lemmas'
r = requests.get(url=url)
```

```python
print(f'code: {r.status_code}. Time: {r.elapsed.total_seconds()}\n')
```

```python
print(f'json: {json.dumps(r.json(), sort_keys=True, indent=1)}')
```

```python
r.json()['data'][0]
```

Obtener la información de una palabra

```python
url = f'{API_URL}/lemmas/ábaco'
r = requests.get(url=url)
```

```python
print(f'code: {r.status_code}. Time: {r.elapsed.total_seconds()}\n')
```

```python
print(f'json: {json.dumps(r.json(), sort_keys=True, indent=1)}')
```

```python
r.json()['data']['pictos'], r.json()['data']['next_lemma']
```
